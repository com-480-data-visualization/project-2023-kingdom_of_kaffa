```python
import matplotlib.pyplot as plt
import pandas as pd

# loading the data
df = pd.read_csv('../dataset/kofio_dataset/kofio_dataset.csv',sep='\t')

df.dropna(subset=['Brewing Method', 'Roast Level', 'Roast Type'], inplace=True)

brewing_methods = ['Vacuum Pot',
 'Hario V60',
 'Clever dripper',
 'Espresso',
 'Chemex',
 'Moka pot',
 'Aeropress',
 'French Press'
]

level_list = ['Light to Medium Light', 'Medium to medium dark', 'Omni']

type_list = ['Filter', 'Espresso', 'Omni']

# Data
count = {}
# for method in brewing_methods:
#     count[method] = len(df[df['Brewing Method'].str.contains(method)])
for level in level_list:
    count[level] = len(df[df['Roast Level'].str.contains(level)])
for t in type_list:
    if t == 'Omni':
        count[t+"_type"] = len(df[df['Roast Type'].str.contains(t)])
    else:
        count[t] = len(df[df['Roast Type'].str.contains(t)])

# Create a bar chart
plt.bar(count.keys(), count.values(), width=0.5)
plt.xticks(fontsize=5)

# Add labels and title
# plt.xlabel("Brewing Method")
plt.ylabel("Num of Items")
plt.title("Clustering")

# Show the plot
plt.show()
```
